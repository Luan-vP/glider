```python
# Setup simulation constants
density = 1.2
viscosity = 0.00002

pilot_weight = 68
```

```python
glider_xml = """
        <body name="body" pos="0 0 1" euler="90 0 0">
            <freejoint/>
            <!-- Main Wing -->
            <geom name="main_wing" type="box" size="0.8 0.05 0.1" rgba="0.8 0.8 0.8 1" pos="0 0 0"/>
            <!-- Pilot -->
        </body>
"""
```

```python
xml = f"""
<mujoco>
    <option density="1.2" viscosity="0.00002" wind="95 0 0"/>
    <worldbody>
        <light name="top" pos="0 0 5"/>
        <!-- Body -->
        {glider_xml}
    </worldbody>
</mujoco>

"""


# Make model and data
model = mujoco.MjModel.from_xml_string(xml)
data = mujoco.MjData(model)

# Make renderer, render and show the pixels
renderer = mujoco.Renderer(model)

mujoco.mj_forward(model, data)
renderer.update_scene(data)

media.show_image(renderer.render())
```

```python
duration = 3.8  # (seconds)
framerate = 60  # (Hz)

# Simulate and display video.
frames = []
mujoco.mj_resetData(model, data)  # Reset state and time.
while data.time < duration:
  mujoco.mj_step(model, data)
  if len(frames) < data.time * framerate:
    renderer.update_scene(data)
    pixels = renderer.render()
    frames.append(pixels)
media.show_video(frames, fps=framerate)
```
